```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Fertilizer Prediction.csv");
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_Soil=LabelEncoder()
df["Soil Type"]=le_Soil.fit_transform(df["Soil Type"])
df["Soil Type"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_Crop=LabelEncoder()
df["Crop Type"]=le_Soil.fit_transform(df["Crop Type"])
df["Crop Type"].unique()
```

```python

X = df.drop("Fertilizer Name", axis=1)
y = df["Fertilizer Name"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model123= RandomForestClassifier();
model123.fit(X_train, y_train)
y_pred = model123.predict(X_test)
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
model12=LogisticRegression()

model12.fit(X_train,y_train)

y_pred12=model12.predict(X_test)
score2=accuracy_score(y_test,y_pred12)
score2
```
